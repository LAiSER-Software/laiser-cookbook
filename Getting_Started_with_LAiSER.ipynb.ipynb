## Import necessary libraries
Here, we are loading the LAiSER library. This contains all the tools we need to extract skills from text.


```python
from laiser.skill_extractor import Skill_Extractor
import pandas as pd
import torch
```

## Sample Description
We are giving the system a short example of a course or job description.
You can change this text to test your own content later.


```python
# Import the dataset
job_sample = pd.read_csv('https://raw.githubusercontent.com/LAiSER-Software/datasets/refs/heads/master/jobs-data/linkedin_jobs_sample_36rows.csv')

job_sample = job_sample[0:1]
job_sample = job_sample[['description', 'job_id']]
print("Considering", len(job_sample), "rows for processing...")
```

##  Initialize the Skill Extractor
Sets up the Skill Extractor, which is the main engine behind the LAiSER library.

```python
print('Initializing the Skill Extractor...')
se = Skill_Extractor(AI_MODEL_ID="marcsun13/gemma-2-9b-it-GPTQ", HF_TOKEN="<YOUR_HUGGING_FACE_API_TOKEN>", use_gpu=True)
print('The Skill Extractor has been initialized successfully!')
```

## Extracting Skills
Now we use LAiSER to extract skills from the description above.
The system reads the description and tries to find related skills using AI.


```python
output = se.extractor(job_sample, 'job_id', text_columns = ['description'])
```

##Viewing Results
Now let's see the extracted skills.
These are based on global skill taxonomies like ESCO and show what a student might learn from the course.

```python
# save the extracted skills to a csv file
display(output)
output.to_csv('extracted_skills_for_sample_jobs.csv', index=False)
```
